## 선형대수학 기초
- 선형대수학(linear algebra)은 벡터(vector)와 행렬(matrix), 벡터 공간(vector space), 선형변환 등을 연구하는 수학의 한 분야이다.   
- 임베딩은 자연어를 벡터로 변환한 결과 혹은 그 과정을 의미하는데, 선형대수학에서 다루는 개념들과 밀접한 관련을 맺고 있다.


### 벡터, 행렬 연산
- 자연어 처리에서는 기본적으로 표와 같은 데이터를 다룬다.   
- 수나 기호, 수식 등을 네모꼴로 배열한 것을 행렬이라고 한다.   
- 벡터는 벡터 공간의 원소다. 나이브하게는 열 또는 행을 하나 떼서 만든 것이라고 생각하면 된다.   
- 벡터는 스칼라 곱, 덧셈, 내적 등 연산이 가능하다.      
- 정방행렬 : 행 개수와 열 개수가 같은 행렬   
- 대칭행렬 : 원래 행렬과 전치가 같은 행렬   
- 단위행렬 : 주대각선의 원소가 모두 1이며 나머지 원소는 모두 0인 정방행렬   
- 역행렬 : 원래 행렬과 곱한 결과가 단위행렬인 행렬을 가리킨다.   

<br>

### 내적과 공분산
- 변수가 여러 개인 다변량 데이터에서는 변수 간 관련성, 즉 상관성이 매우 중요한 문제가 된다.   
- 상관계수는 두 확률변수의 직선 관계가 얼마나 강하고 어떤 방향인지를 나타내는 값이라고 볼 수 있다.   
- 공분산행렬은 각 요소가 그에 해당하는 변수 간 공분산인 행렬을 의미한다.   
- 내적 (Dot Product): 내적은 두 벡터 간의 곱셈 연산입니다.      
- 공분산 (Covariance): 공분산은 두 변수 간의 관계를 나타냅니다.   

<br>

### 행렬 분해 기반 차원 축소(1): 주성분 분석 (PCA)
- 차원축소(dimensinality reduction) 기법은 데이터의 차원을 줄여 처리를 효율화하는 한편 시각화하는 데 쓰인다.   
- 차원 축소 기법 가운데 널리 쓰이는 방법이 행렬분해(matric factorization) 기반의 방법들이다.   
- 대표적으로 주성분 분석과 특이값 분해가 있다.

- PCA는 데이터의 분산을 최대한 보존하면서 서로 직교하는 새 기저(축)을 찾아,   
고차원 공간의 표본들을 선형 연관성이 없는 저차원 공간으로 변환하는 기법이다.   

<br>

### 행렬 분해 기반 차원 축소(2): 특이값 분해(SVD)
- 특이값 분해 (Singular Value Decomposition)
- SVD는 데이터를 더 작고 의미 있는 부분으로 분해하고, PCA는 데이터의 주요 특성을 찾아내어 데이터를 더 간단하게 만듭니다.
