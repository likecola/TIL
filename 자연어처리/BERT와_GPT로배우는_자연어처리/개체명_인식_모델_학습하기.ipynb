# 각종 설정
모델 하이퍼파라메터(hyperparameter)와 저장 위치 등 설정 정보를 선언합니다.

```python
import torch
from ratsnlp.nlpbook.ner import NERTrainArguments
args = NERTrainArguments(
    pretrained_model_name="beomi/kcbert-base",
    downstream_corpus_name="ner",
    downstream_model_dir="/gdrive/My Drive/nlpbook/checkpoint-ner",
    batch_size=32 if torch.cuda.is_available() else 4,
    learning_rate=5e-5,
    max_seq_length=64,
    epochs=3,
    tpu_cores=0 if torch.cuda.is_available() else 8,
    seed=7,
)
```

# 랜덤 시드 고정
학습 재현을 위해 랜덤 시드를 고정합니다.

```python
from ratsnlp import nlpbook
nlpbook.set_seed(args)
```

# 로거 설정
메세지 출력 등을 위한 logger를 설정합니다.

```python
nlpbook.set_logger(args)
```

# 말뭉치 다운로드
실습에 사용할 말뭉치(NER)를 다운로드합니다.

```python
nlpbook.download_downstream_dataset(args)
```

# 토크나이저 준비
토큰화를 수행하는 토크나이저를 선언합니다

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained(
    args.pretrained_model_name,
    do_lower_case=False,
)
```

# 학습데이터 구축
학습데이터를 만듭니다.

```python
from ratsnlp.nlpbook.ner import NERCorpus, NERDataset
from torch.utils.data import DataLoader, SequentialSampler, RandomSampler
corpus = NERCorpus(args)
# NERCorpus 클래스는 텍스트 파일을 '원본 문장 + 개체명 태그를 레이블한 문장' 형태로 읽어들입니다.
# NERDataset 클래스는 NERCorpus와 코드에서 선언해 둔 토크나이저를 품고있습니다.
train_dataset = NERDataset(
    args=args,
    corpus=corpus,
    tokenizer=tokenizer,
    mode="train",
)
train_dataloader = DataLoader(
    train_dataset,
    batch_size=args.batch_size,
    sampler=RandomSampler(train_dataset, replacement=False),
    collate_fn=nlpbook.data_collator,
    drop_last=False,
    num_workers=args.cpu_workers,
)
```

```python
train_dataset[3]
```

NERFeatures 라는 자료형에는 네 가지 정보가 있습니다.
1. input_ids : 인덱스로 변환된 토큰 시퀀스입니다.
2. attention_mask : 해당 토큰이 패딩 토큰인지 (0) 아닌지(1)를 나타냅니다.
3. token_type_ids : 세그먼트
4. label_ids : 정수로 바뀐 레이블 시퀀스입니다.

# 테스트 데이터 구축
학습 중에 평가할 테스트 데이터를 구축합니다.

```python
val_dataset = NERDataset(
    args=args,
    corpus=corpus,
    tokenizer=tokenizer,
    mode="val",
)
val_dataloader = DataLoader(
    val_dataset,
    batch_size=args.batch_size,
    sampler=SequentialSampler(val_dataset),
    collate_fn=nlpbook.data_collator,
    drop_last=False,
    num_workers=args.cpu_workers,
)
```

# 모델 초기화
프리트레인이 완료된 BERT 모델을 읽고, 개체명 인식 수행할 모델을 초기화합니다.

```python
from transformers import BertConfig
pretrained_model_config = BertConfig.from_pretrained(
    args.pretrained_model_name,
    num_labels=corpus.num_labels,
)
```

```python
from transformers import BertForTokenClassification
model = BertForTokenClassification.from_pretrained(
        args.pretrained_model_name,
        config=pretrained_model_config,
)
```

# 학습 준비
Task와 Trainer를 준비합니다.

```python
from ratsnlp.nlpbook.ner import NERTask
task = NERTask(model, args)
```

```python
trainer = nlpbook.get_trainer(args)
```

# 학습
준비한 데이터와 모델로 학습을 시작합니다. 학습 결과물(체크포인트)은 미리 연동해둔 구글 드라이브의 준비된 위치(`/gdrive/My Drive/nlpbook/checkpoint-ner`)에 저장됩니다.

```python
trainer.fit(
    task,
    train_dataloaders=train_dataloader,
    val_dataloaders=val_dataloader,
)
```
